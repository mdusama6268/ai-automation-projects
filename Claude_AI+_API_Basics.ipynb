````python
import json

# Clean and parse the response
# Find the start and end of the JSON object
json_start = response.find('{')
json_end = response.rfind('}')

if json_start != -1 and json_end != -1:
    clean = response[json_start : json_end + 1]
else:
    # Fallback to original stripping if JSON markers are not found, though this might still cause errors if AI deviates
    clean = response.strip("```json").strip("```").strip()

data = json.loads(clean)

# Process each classification automatically
print("--- AUTO ROUTING SYSTEM ---\n")

for item in data["classifications"]:
    message = item["message"]
    category = item["category"]

    # Auto route to correct team
    if category == "DELAY":
        team = "Logistics Team"
        action = "Check driver location"
    elif category == "WRONG_ORDER":
        team = "Restaurant Team"
        action = "Verify order with restaurant"
    elif category == "REFUND":
        team = "Finance Team"
        action = "Process refund within 24 hours"
    elif category == "COMPLAINT":
        team = "Quality Team"
        action = "Escalate to quality manager"

    print(f"Message: {message}")
    print(f"Category: {category}")
    print(f"Assigned to: {team}")
    print(f"Action: {action}")
    print("-" * 40)
````

````python
from groq import Groq
from google.colab import userdata

client = Groq(api_key=userdata.get("GROQ_API_KEY"))

print("Ready!")
````

````python
# conversation history list — this is the memory!
conversation_history = [
    {
        "role": "system",
        "content": """You are a helpful AI assistant for FoodRush
— a food delivery company in Bhopal.
You help operations managers with:
- Delivery tracking and updates
- Driver management
- Customer complaints
- Order analytics
Always be professional and concise."""
    }
]

def chat(user_message):
    # Step 1 — Add user message to history
    conversation_history.append({
        "role": "user",
        "content": user_message
    })

    # Step 2 — Send full history to AI
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=conversation_history
    )

    # Step 3 — Get AI reply
    ai_reply = response.choices[0].message.content

    # Step 4 — Add AI reply to history too
    conversation_history.append({
        "role": "assistant",
        "content": ai_reply
    })

    return ai_reply

print("Chatbot Ready!")
````

````python
import PyPDF2

def extract_pdf_text(filename):
    text = ""
    with open(filename, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        total_pages = len(reader.pages)
        print(f"Total Pages: {total_pages}")

        for page_num in range(total_pages):
            page = reader.pages[page_num]
            text += f"\n--- Page {page_num + 1} ---\n"
            text += page.extract_text()

    return text

# Extract text
pdf_text = extract_pdf_text(filename)

# Show first 500 characters to confirm it worked
print("\nFirst 500 characters extracted:")
print("-" * 40)
print(pdf_text[:500])
print("-" * 40)
print(f"\nTotal characters extracted: {len(pdf_text)}")
````

````python
from groq import Groq
from google.colab import userdata, files
import PyPDF2
import requests
import json

client = Groq(api_key=userdata.get("GROQ_API_KEY"))

print("=" * 55)
print("   FoodRush AI Operations Command Center")
print("   Initializing systems...")
print("=" * 55)
````

````python
# Function 1 — Ask AI
def ask_ai(system_prompt, user_message):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_message}
        ]
    )
    return response.choices[0].message.content

# Function 2 — Read PDF
def read_pdf(filename):
    text = ""
    with open(filename, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        print(f"PDF loaded — {len(reader.pages)} pages found")
        for page in reader.pages:
            text += page.extract_text()
    return text

# Function 3 — Get Weather
def get_weather(city):
    api_key = "bd5e378503939ddaee76f12ad7a97608"
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    data = requests.get(url).json()
    return {
        "city": data["name"],
        "temp": data["main"]["temp"],
        "condition": data["weather"][0]["description"],
        "humidity": data["main"]["humidity"],
        "wind": data["wind"]["speed"]
    }

# Function 4 — Classify Complaint
def classify_complaint(complaint):
    result = ask_ai(
        system_prompt="You are a complaint classifier. Return ONLY one word: DELAY, WRONG_ORDER, REFUND, or COMPLAINT",
        user_message=complaint
    )
    return result.strip()

print("All functions loaded successfully!")
````

````python
# Step 1 — Get live weather
print("Fetching live weather for Bhopal...")
weather = get_weather("Bhopal")
print(f"Weather fetched — {weather['temp']}°C {weather['condition']}")

# Step 2 — Generate AI morning briefing
print("\nGenerating morning briefing...")

briefing_response = ask_ai(
    system_prompt="""You are an expert operations manager
for FoodRush — a food delivery company in Bhopal.
Generate a professional morning briefing report.""",

    user_message=f"""Generate a complete morning operations
briefing based on this data:

Weather in {weather['city']}:
- Temperature: {weather['temp']}°C
- Condition: {weather['condition']}
- Humidity: {weather['humidity']}%
- Wind Speed: {weather['wind']} m/s

Return as JSON:
{{
  "date": "Today",
  "weather_impact": "...",
  "high_demand_items": ["...", "...", "..."],
  "driver_advisory": ["...", "...", "..."],
  "promotions": ["...", "...", "..."],
  "priority_action": "..."
}}"""
)

# Step 3 — Parse and display briefing
clean = briefing_response.strip("```json").strip("```").strip()
briefing = json.loads(clean)

print("\n" + "=" * 55)
print("   🌅 FOODRUSH MORNING BRIEFING — BHOPAL")
print("=" * 55)
print(f"\n🌤️  WEATHER: {weather['temp']}°C — {weather['condition']}")
print(f"\n📊 WEATHER IMPACT:")
print(f"   {briefing['weather_impact']}")
print(f"\n🍦 HIGH DEMAND ITEMS:")
for item in briefing["high_demand_items"]:
    print(f"   • {item}")
print(f"\n🚴 DRIVER ADVISORY:")
for tip in briefing["driver_advisory"]:
    print(f"   • {tip}")
print(f"\n🎯 PROMOTIONS TO RUN:")
for promo in briefing["promotions"]:
    print(f"   • {promo}")
print(f"\n⚡ PRIORITY ACTION:")
print(f"   {briefing['priority_action']}")
print("\n" + "=" * 55)
````

````python
# Simulate incoming customer complaints
complaints = [
    "My order from Bhopal Biryani House is 1 hour late!",
    "I ordered paneer tikka but received chicken tikka",
    "The food was completely cold and stale",
    "Please refund my money for order #BPL-2341",
    "Driver was very rude and threw the food bag"
]

print("=" * 55)
print("   📨 COMPLAINT AUTO ROUTING SYSTEM")
print("=" * 55)
print()

# Route each complaint
routing_results = []

for complaint in complaints:
    category = classify_complaint(complaint)

    # Assign team and action
    if "DELAY" in category:
        team = "Logistics Team"
        action = "Check driver GPS location"
        priority = "🔴 HIGH"
    elif "WRONG_ORDER" in category:
        team = "Restaurant Team"
        action = "Verify order with restaurant"
        priority = "🟡 MEDIUM"
    elif "REFUND" in category:
        team = "Finance Team"
        action = "Process refund in 24 hours"
        priority = "🟡 MEDIUM"
    elif "COMPLAINT" in category:
        team = "Quality Team"
        action = "Escalate to quality manager"
        priority = "🔴 HIGH"
    else:
        team = "Support Team"
        action = "Review manually"
        priority = "🟢 LOW"

    routing_results.append({
        "complaint": complaint,
        "category": category,
        "team": team,
        "action": action,
        "priority": priority
    })

    print(f"{priority} | {category}")
    print(f"Complaint : {complaint}")
    print(f"Assigned  : {team}")
    print(f"Action    : {action}")
    print("-" * 55)

print(f"\nTotal complaints processed: {len(complaints)}")
print("All complaints auto routed successfully!")
````

````python
# Build full conversation history
command_center_history = [
    {
        "role": "system",
        "content": f"""You are the FoodRush AI Operations
Command Center for Bhopal.

You have access to:

LIVE WEATHER DATA:
- City: {weather['city']}
- Temperature: {weather['temp']}°C
- Condition: {weather['condition']}
- Humidity: {weather['humidity']}%
- Wind: {weather['wind']} m/s

TODAY'S COMPLAINTS PROCESSED:
{json.dumps(routing_results, indent=2)}

MORNING BRIEFING:
- Weather Impact: {briefing['weather_impact']}
- Priority Action: {briefing['priority_action']}

Answer all operations questions using this data.
Be concise, professional, and actionable.
Always relate answers to Bhopal food delivery context."""
    }
]

def command_center_chat(message):
    command_center_history.append({
        "role": "user",
        "content": message
    })

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=command_center_history
    )

    reply = response.choices[0].message.content

    command_center_history.append({
        "role": "assistant",
        "content": reply
    })

    return reply

# Exit words
exit_words = ["quit", "bye", "exit", "stop"]

print("=" * 55)
print("   🚀 FOODRUSH AI COMMAND CENTER — LIVE")
print("   Ask anything about today's operations!")
print("   Type 'quit' to exit")
print("=" * 55)
print()

while True:
    user_input = input("You: ")

    if user_input.lower().strip() in exit_words:
        print("AI: Command Center shutting down. Good luck today!")
        break

    if user_input.strip() == "":
        continue

    response = command_center_chat(user_input)
    print(f"\nAI: {response}\n")
````
